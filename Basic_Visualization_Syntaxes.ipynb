```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.shape
```

```python
df.total_bill.plot.hist()
```

```python
sns.distplot(df.total_bill)
```

```python
sns.boxplot(df.total_bill)
```

```python
sns.scatterplot(df.total_bill,df.tip)
```

```python
df.total_bill.plot()
```

```python
sns.barplot(x="day", y="tip", data=df)
```

```python
sns.barplot(x="tip", y="day", data=df)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="day", y="tip", hue="sex", data=df)
```

```python
sns.countplot(x="day", hue="sex", data=df)
```

```python
# box plot
```

```python
sns.boxplot(x="day", y="total_bill", data=df)
```

```python
sns.scatterplot(df.total_bill,df.tip,hue=df.day)
```

```python
sns.violinplot(x="day", y="total_bill", data=df)
```
